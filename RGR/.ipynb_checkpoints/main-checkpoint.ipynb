```python
import numpy as np
from random import randint
from math import sqrt

def steepest(A, b, x, eps, maxiter):
    AT = np.transpose(A)
    A = np.matmul(AT, A)
    b = np.matmul(AT, b)
    
    i = 1
    r = b - np.matmul(A, x)
    delta = np.dot(r, r)
    delta0 = delta
    conv = [delta]
    while (delta > eps * delta0 and i < maxiter):
        q = np.matmul(A, r)
        alpha = delta / np.dot(q, r)
        x = x + alpha * r
        if (i % 500 == 0):
            r = b - np.matmul(A, x)
        else:
            r = r - alpha * q
        delta = np.dot(r, r)
        i += 1
        conv.append(delta)
    return x, conv
```

```python
n = 100
A = [[randint(1, 100) for k in range(n)] for i in range(n)]
X = [randint(1, 100) for i in range(n)]
f = np.matmul(A, X)
x0 = [0 for _ in range(n)]

solution = steepest(A, f, x0, 0.00001, 1000000)
```

```python
import matplotlib.pyplot as plt

plt.plot([i for i in range(len(solution[1]))], solution[1])
plt.grid()
plt.yscale('log')
print(X)
print(solution[0])
```
